#K Nearest Neighbors
####Task: predict the label for a given sample

**Step 1**: find the k nearest neighbors of the sample
1. calculate the distance to every know sample
2. take the nearest k samples

**Step 2**: predict the lable
1. find the most popular class of the k nearest samples
2. return it as the prediction

##1. Please complete the Sample class

```python
class Sample:
  def __init__(self, features:list, label:str):
    self.features = features
    self.label = label

  def get_features(self):
    return self.features

  def set_features(self, features:list):
    self.features = features

  def get_label(self):
    return self.label

  def set_label(self, label:str):
    self.label = label

  def distance(self, sample):
    '''
    calculate the distance between two samples
    '''
    dist = 0
    feats = sample.get_features()
    for i in range(len(feats)):
      dist += (self.features[i] - feats[i])**2
    return dist**0.5


#test
sample1 = Sample([1,2,3], 'A')
sample2 = Sample([4,5,6], 'B')
print(sample1.distance(sample2))
```

#2. Please complete the KnnPredictor

```python
class KnnPredictor:

  def __init__(self, samples:list, k:int):
    self.samples = samples
    self.k = k

  def find_k_neighbors(self, new_sample):
    '''
    1. calculate the distance to every know sample
    2. return the nearest k samples of the new sample
    '''
    dist = {}

    for s in self.samples:
      d = s.distance(new_sample)
      try:
        dist[d].append(s)
      except KeyError:
        dist[d] = [s]
    sortedKey = sorted(dist)
    sortedSamples = []
    for key in sortedKey:
      sortedSamples.extend(dist[key])

    return sortedSamples[:self.k]


  def find_the_most_popular_class(self, neighbors):
    '''
    find the most popular class of the k nearest samples
    return it as the prediction
    '''
    votes = {}
    """
    votes = {
        "0": the number of the neightbors of "0" label
        "1": the number of the neightbors of "1" label
        }
    """
    for s in neighbors:
        l = s.get_label()
        try:
            votes[l] += 1
        except KeyError:
            votes[l] = 1
    max_label = ""
    max_count = 0
    for key, value in votes.items():
        if value > max_count:
            max_label = key
            max_count = value
    return max_label

  def predict(self, sample):
    neighbors = self.find_k_neighbors(sample)
    prediction = self.find_the_most_popular_class(neighbors)
    return prediction

#test
samples = [Sample([1,2,3], 'A'), Sample([4,5,6], 'B'), Sample([7,8,9], 'C')]
knn = KnnPredictor(samples, 2)
print(knn.predict(Sample([10,11,12], 'Unknown')))
```
